What are lists?

Lists is a data type where you can store multiple items under 1 name. More technically, lists act like dynamic arrays which means u can add more items to a fly



Why lists are required in programming?

Lists are mutable, which means that you can change the elements of a list.

A list, sometimes called an array, is a tool for storing data, just like a variable. However, unlike a variable, which can only store one piece of data, a list can store many. Like a variable, a list is a named storage area within the computer's memory.

Nested if with list comprehension

```python
basket = ['apple','guava','cherry','banana']

my_fruits = ['apple','kiwi']

# This is a list comprehension that creates a new list containing all the fruits from `my_fruits` that are also present in `basket` and start with the letter 'a'.

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

Cartsian product using list comprihension

```python
list_a = [1, 2, 3]
list_b = [4, 5, 6]

cartesian_product = [(a, b) for a in list_a for b in list_b]

print(cartesian_product)
```

SETS

A set is an unordered collection of items. Every set element is unique and must be immutable(cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like unique, intersection, symmetric diffrence, etc
CHARECTRESTICS
1. Uordered
2. Mutable
3. No Duplications
4. Can't contain mutable data types

```python
class Student:
    def __init__(self, name, age, roll_no):
        self.name = name
        self.age = age
        self.roll_no = roll_no

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll No:", self.roll_no)


class EngineeringStudent(Student):
    def __init__(self, name, age, roll_no, branch):
        super().__init__(name, age, roll_no)
        self.branch = branch

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll No:", self.roll_no)
        print("Branch:", self.branch)


class MedicalStudent(Student):
    def __init__(self, name, age, roll_no, specialization):
        super().__init__(name, age, roll_no)
        self.specialization = specialization

    def display_details(self):
        print("Name:", self.name)
        print("Age:", self.age)
        print("Roll No:", self.roll_no)
        print("Specialization:", self.specialization)


def create_student():
    student_type = input("Enter student type (E - Engineering, M - Medical): ")
    name = input("Name: ")
    age = int(input("Age: "))
    roll_no = int(input("Roll No: "))

    if student_type == "E":
        branch = input("Enter branch: ")
        student = EngineeringStudent(name, age, roll_no, branch)
    elif student_type == "M":
        specialization = input("Enter specialization: ")
        student = MedicalStudent(name, age, roll_no, specialization)
    else:
        print("Invalid student type.")
        return None

    return student


def main():
    student = create_student()
    if student:
        student.display_details()


if __name__ == "__main__":
    main()

```
